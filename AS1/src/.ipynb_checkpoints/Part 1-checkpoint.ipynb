```python
from sklearn.metrics.pairwise import pairwise_distances
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.utils import shuffle
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold

from model import *
```

```python
# Read each of the single var datasets
sets = []
for i in range(1,5):
    sets.append(pd.read_csv(f"http://www.cs.iit.edu/~agam/cs584/share/svar-set{i}.dat",sep=' ',skiprows=5, names=['x','y']).reset_index(drop=True))
```

```python
def plot_data(x,y,title):
    plt.scatter(x,y)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.show()
    
for i,data in enumerate(sets):
    plot_data(data.x.values,data.y.values,f"Single Var Set {i}")
```

```python
# Fit linear regression model on train test split and compare train test error

for i,data in enumerate(sets):
    # Split into training and testing set
    X = data.x.values.reshape((data.shape[0],-1))
    y = data.y.values
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
    # Fit linear model on train data
    lr = LinearRegressionExplicit()
    lr.fit(X_train,y_train)
    
    # Predict on train and test
    train_pred = lr.predict(X_train)
    test_pred = lr.predict(X_test)
    
    # Train and test error
    train_err = regression_error(train_pred,y_train)
    test_err = regression_error(test_pred,y_test)
    
    print("="*40)
    print(f"Dataset {i}: \n\tTrain error: {train_err}\n\tTest error: {test_err}")
```

```python
# Fit linear regression model on the whole data and plot the model
for i,data in enumerate(sets):
    # Fit the model on the whole data
    X = data.x.values.reshape((data.shape[0],-1))
    y = data.y.values
    lr = LinearRegressionExplicit()
    lr.fit(X,y)
    pred = lr.predict(X)
    
    # Plot model
    plt.scatter(X,y,color='magenta')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.plot(X,pred,color='black',linewidth=2)
    plt.title(f"Model on dataset {i}")
    plt.show()
```

```python
# Perform KFold cross validation to choose polynomial model for each
kf = KFold(n_splits=5)
degrees = [2,3,4,5]
best_models = []
for i, data in enumerate(sets[1:]):
    X = data.x.values.reshape((data.shape[0],-1))
    y = data.y.values
    average_errors = []
    for degree in degrees:
        all_errors = []
        for train_index, test_index in kf.split(X):
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]
            
            # Create polynomial features
            poly_reg = PolynomialFeatures(degree=degree)
            X_train_poly = poly_reg.fit_transform(X_train)
            X_test_poly = poly_reg.fit_transform(X_test)
            
            # Train LR
            lr = LinearRegressionExplicit()
            lr.fit(X_train_poly,y_train)
            
            # Predict and error
            test_pred = lr.predict(X_test_poly)
            all_errors.append(regression_error(test_pred,y_test))
        average_errors.append(np.average(all_errors))
    
    print(f"Dataset {i+1}")
    for i,average_error in enumerate(average_errors):
        print(f"\tTesting error for degree {degrees[i]}: {average_error}")
    print(f"\tPolynomial of degree {degrees[np.argmin(average_errors)]} gives smallest test error")
    
        
```

```python
# Fit polynomial regression model on the whole data and plot the model
for i,data in enumerate(sets[1:]):
    
    # Fit the model on the whole data
    X = data.x.values.reshape((data.shape[0],-1))
    y = data.y.values
    
    if i+1 == 1:
        degree = 3
    else:
        degree = 4
    poly_reg = PolynomialFeatures(degree=degree)
    X_poly = poly_reg.fit_transform(X)
    lr = LinearRegressionExplicit()
    lr.fit(X_poly,y)
    pred = lr.predict(X_poly)
    
    # Plot model
    plt.scatter(X,y,color='magenta')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.scatter(X,pred,color='black',linewidth=2)
    plt.title(f"Model on dataset {i+2}")
    plt.show()
```
